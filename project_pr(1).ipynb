```python
import pandas as pd
import numpy as np

data = pd.read_csv('voice.csv',decimal='.')
print data[:3]
```

```python
data.describe()
```

```python
%matplotlib inline
data.hist(figsize = (20,50), bins = 200)
```

```python
import matplotlib.pyplot as plt

x = pd.DataFrame(data[:1500],columns=['IQR'])
y = pd.DataFrame(data[:1500],columns=['meanfun'])

x1 = pd.DataFrame(data[1600:3100],columns=['IQR'])
y1 = pd.DataFrame(data[1600:3100],columns=['meanfun'])

color = 'red'
    
plt.scatter(x,y)
plt.show()

plt.scatter(x1,y1,c = color)
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x, y, s=10, c='b', marker="s", label='male')
ax1.scatter(x1, y1, s=10, c='r', marker="o", label='female')
plt.legend(loc='upper right');
plt.xlabel('IQR')
plt.ylabel('meanfun')
plt.show()
```
